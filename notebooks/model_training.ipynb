```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("./data/gemstone_train.csv")  # here . represents the current directory

data
```

```python
# dropping id column from gemstone dataframe

data.drop(labels=["id"], axis=1, inplace=True)

data
```

```python
# segregating independent (X) and dependent (y) features
# here price is dependent feature and rest are independent feature

X=data.drop(["price"], axis=1)

y=data[["price"]]
```

```python
# Independent features

X
```

```python
# Dependent feature

y
```

## Segregating categorical and numerical features from the Independent features dataframe

```python
# calling columns of the numerical features

X.select_dtypes(exclude="object").columns
```

```python
# calling columns of the categorical features

X.select_dtypes(include="object").columns
```

```python
# numerical features

num_feature = X.select_dtypes(exclude="object")
num_feature
```

```python
# categorical features

cat_feature = X.select_dtypes(include="object")
cat_feature
```

```python
# Defining the custom ranking for each ordinal variable in the categorical features

cut_categories = ['Fair', 'Good', 'Very Good','Premium','Ideal']

color_categories = ['J', 'I', 'H', 'G', 'F', 'E', 'D']

clarity_categories = ['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']
```

```python
# for handling missing values
from sklearn.impute import SimpleImputer

# for scaling data into standard normal distribution
from sklearn.preprocessing import StandardScaler

# for encoding ordinal categorical features
from sklearn.preprocessing import OrdinalEncoder

# for pipeline
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
# creating a pipeline for numerical feature

num_pipeline = Pipeline(
    steps=[
        ("imputer", SimpleImputer()),
        ("scaler", StandardScaler())
    ]
)
```

```python
# creating pipeline for categorical features

cat_pipeline = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("ordinal", OrdinalEncoder(categories=[cut_categories, color_categories, clarity_categories]))
    ]
)
```

```python
# transforming the pipelines created

preprocessor = ColumnTransformer(
    [
        ("num", num_pipeline, num_feature.columns),
        ("cat", cat_pipeline, cat_feature.columns)
    ]
)
```

```python
# dividing the independent (X) and dependent (y) features into train and test data

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# my X_train data before preprocessing

X_train
```

```python
# my X_test data before preprocessing

X_test
```

```python
# initializing preprocessing on independent train and test data

# fit_transform on X_train data
preprocessor.fit_transform(X_train)
```

```python
# applying transform on X_test data
preprocessor.transform(X_test)
```

```python
# getting the names of the columns of the transformed dataset

preprocessor.get_feature_names_out()
```

```python
# converting X_train data and X_test data into dataframe

# X_train
X_train = pd.DataFrame(data=preprocessor.fit_transform(X_train), columns=preprocessor.get_feature_names_out())

# X_test
X_test = pd.DataFrame(data=preprocessor.transform(X_test), columns=preprocessor.get_feature_names_out())
```

```python
# X_train data after preprocessing

X_train
```

```python
# X_test data after preprocessing

X_test
```

## Model training

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
# creating a function for calculating R-Squared stats, MSE, RMSE and MAE

def evaluate_model(true, predicted):
    R2_score = r2_score(true, predicted)
    MSE = mean_squared_error(true, predicted)
    RMSE = np.sqrt(MSE)
    MAE = mean_absolute_error(true, predicted)

    return RMSE, MAE, R2_score
```

```python
# creating dictionary of different models for looping purpose

models = {
    "LinearRegression Model": LinearRegression(),
    "Ridge Model": Ridge(),
    "Lasso Model": Lasso(),
    "ElasticNet Model": ElasticNet()
}
```

```python
# fetching keys of model dictionary

models.keys()
```

```python
# fetching values of model dictionary

models.values()
```

```python
# looping values (i.e. models) from models dictionary

for i in range(len(models)):

    model = list(models.values())[i]
    print(model)
```

```python
# training the model and then predicting using test data

for i in range(len(models)):

    # fetching model
    model=list(models.values())[i]

    # training the model
    model.fit(X_train, y_train)

    # model is trained
    # making predictions now
    y_pred = model.predict(X_test)

    # checking validation (test) score of the model
    RMSE, MAE, R2_score = evaluate_model(y_test, y_pred)

    print(str(i+1)+".",  list(models.keys())[i])

    print("Training Performance of the Model")
    print(f"root_mean_squared_error: {RMSE}")
    print(f"mean_absolute_error: {MAE}")
    print(f"r2_score: {R2_score*100}")

    print("\n")
```
